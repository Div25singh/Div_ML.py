```python
from __future__ import print_function
import os
data_path = ['data']
print(data_path)
```

```python
import numpy as np
import pandas as pd
import os
filepath = os.sep.join(data_path + ['irisd.csv'])
print(filepath)
data=pd.read_csv(filepath)
print(data.head())
```

```python
#plot histogram
plt.hist(data.sepal_length,bins=25)
```

```python
import seaborn as sns
sns.set_context('notebook')

ax= data.plot.hist(bins=25,alpha=0.5)
ax.set_xlabel('size(cm)');
```

```python
#to create four seperate plots, use pandas , hist
axlist =data.hist(bins=25)

for ax in axlist.flatten():
    if ax.is_last_row():
        ax.set_xlabel('size(cm)')
        
    if ax.is_first_col():
        ax.set_ylabel('Frequency')
```
