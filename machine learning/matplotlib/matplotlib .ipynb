# Matplotlib
Matplotlib is a library for making 2D plots of arrays in Python.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(start=0,stop=100,step=5)
y = x**2
```

```python
# line plotting
plt.plot(x,y,'r--')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.show()
```

```python
x = np.arange(start=0,stop=10,step=2)
y = x**2
```

```python
# line plotting
plt.plot(x,y,'r--')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.show()
```

```python
x = np.arange(start=0,stop=10,step=1)
y = x**2
```

```python
# line plotting
plt.plot(x,y,'r--')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.show()
```

```python
# ploting sine wave
x = np.arange(0, 10, 0.5);
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
# data point plotting
data_x = [1,2,3,4,5,6]
data_y = [1,1,2,3,5,8]
plt.plot(data_x, data_y, 'ro')
plt.axis([0, 7, 0, 10])
plt.show()
```

```python
# data point plotting
data_x = [1,2,3,4,5,6]
data_y = [4,1,2,10,5,7]
plt.plot(data_x, data_y, 'ro')
plt.axis([0, 7, 0, 10])
plt.show()
```

```python
# simple bar chart 
x_label = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
x_position = np.arange(len(x_label))
data_y = [10,8,6,4,2,1]
plt.bar(x_position, data_y, align='center', alpha=1)
plt.xticks(x_position, x_label)
plt.title('Programming languages')
plt.show()
```

```python
# pie chart
label = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
ratio = [10,8,6,4,2,1]
plt.pie(ratio, labels = label, startangle=180)
plt.axis('equal')
plt.show()
```
